## Extract and fit peri-saccadic probe responses

```python
from myphdlib.figures.modulation import BasicSaccadicModulationAnalysis
mod = BasicSaccadicModulationAnalysis(
    hdf='/media/jbhunt/EPHYS-01-BACKUP/opto/NOPEmod.hdf',
    tag='EPHYS-01-BACKUP',
    experiment=('NOPETangential')
)
mod.loadNamespace()
mod.computeSaccadeResponseTemplates()
mod.computePerisaccadicPeths()
mod.fitPerisaccadicPeths()
mod.saveNamespace()
```

## Compute p-values for each MI estimate

```python
from myphdlib.figures.bootstrap import BootstrappedSaccadicModulationAnalysis
boot = BootstrappedSaccadicModulationAnalysis(
    hdf='/media/jbhunt/EPHYS-01-BACKUP/opto/NOPEmod.hdf',
    tag='EPHYS-01-BACKUP',
    experiment=('NOPETangential')
)
boot.loadNamespace()
boot.resampleExtrasaccadicPeths(nRuns=3000)
boot.generateNullSamples(parallelize=False)
boot.computeProbabilityValues()
boot.saveNamespace()
```

## Extract target information
* MI values for the time window immediately after saccades for the largest response component is here: `boot.ns['mi/pref/real'][:, 5, 0]`
* P-values for the time window immediately after saccades for the largest response component is here: `boot.ns['p/pref/real'][:, 5, 0]`
* Unit cluster is the 3rd element of the tuples stored here: `boot.ukeys`
