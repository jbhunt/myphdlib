```python
from myphdlib.general.labjack import filterPulsesFromPhotologicDevice, filterPulsesWithSmoothing
from myphdlib.interface.nope import NOPESession
import numpy as np
```

```python
session = NOPESession('/media/jbhunt/EPHYS-01-BACKUP/opto/2025-09-23/ChR2GAD2')
labjackData = session.load('labjack/matrix')
#defining the labjack data should also be edited to pull from hdf
timestamps = labjackData[:, 0]
TTLdata = labjackData[:, 9]
# minimumPulseWidthInSeconds adjusted from 0.013 to check pulse count mismatch origin
try:    
    filtered = filterPulsesFromPhotologicDevice(labjackData[:, 9],
    minimumPulseWidthInSeconds=0.009)
except:
    print('smoothing')
    filtered = filterPulsesWithSmoothing(labjackData[:, 9])
```
