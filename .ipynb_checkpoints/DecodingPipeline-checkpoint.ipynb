```python
keys = list()
for session in sessionsMlati:
    keys.append(str(session.date))
premotorUnitsZeta = {key:[] for key in keys}
for session in sessionsMlati:
    labels = session.load('nptracer/labels')
    transformed = session.load('nptracer/transformed')
    brainAreas = ul.translateBrainAreaIdentities(labels) 
    spikeClustersFile = session.home.joinpath('ephys/sorting/manual/spike_clusters.npy')
    uniqueSpikeClusters = np.unique(np.load(spikeClustersFile))
    zetaNasal = session.load('zeta/saccade/nasal/p')
    zetaTemporal = session.load('zeta/saccade/temporal/p')
    ampCutoff = session.load('metrics/ac')
    presenceRatio = session.load('metrics/pr')
    firingRate = session.load('metrics/fr')
    isiViol = session.load('metrics/rpvr')
    qualityLabels = session.load('metrics/ql')
    for index, pVal in enumerate(zetaNasal):
        if brainAreas[index] in ['SCsg','SCop', 'SCig', 'SCiw', 'SCdg']:
            pNasal = pVal
            pTemporal = zetaTemporal[index]
            if pNasal < pTemporal:
                p = pNasal
            elif pTemporal < pNasal:
                p = pTemporal
            if p < 0.05:
                if qualityLabels is not None and qualityLabels[index] in (0, 1):
                        continue
                if ampCutoff[index] <= 0.1:
                    if presenceRatio[index] >= 0.9:
                        if firingRate[index] >= 0.2:
                            if isiViol[index] <= 0.5:
                                date = str(session.date)
                                unit = uniqueSpikeClusters[index]
                                premotorUnitsZeta[date].append(unit)
```

```python
keys = list()
for session in sessionsMlati:
    keys.append(str(session.date))
visualUnitsZeta = {key:[] for key in keys}
for session in sessionsMlati:
    labels = session.load('nptracer/labels')
    transformed = session.load('nptracer/transformed')
    spikeClustersFile = session.home.joinpath('ephys/sorting/manual/spike_clusters.npy')
    uniqueSpikeClusters = np.unique(np.load(spikeClustersFile))
    brainAreas = ul.translateBrainAreaIdentities(labels) 
    zetaLeft = session.load('zeta/probe/left/p')
    zetaRight = session.load('zeta/probe/right/p')
    ampCutoff = session.load('metrics/ac')
    presenceRatio = session.load('metrics/pr')
    firingRate = session.load('metrics/fr')
    isiViol = session.load('metrics/rpvr')
    qualityLabels = session.load('metrics/ql')
    for index, pVal in enumerate(zetaLeft):
        if brainAreas[index] in ['SCsg','SCop', 'SCig', 'SCiw', 'SCdg']:
            pLeft = pVal
            pRight = zetaRight[index]
            if pLeft < pRight:
                p = pLeft
            elif pRight < pLeft:
                p = pRight
            if p < 0.05:
                if qualityLabels is not None and qualityLabels[index] in (0, 1):
                        continue
                if ampCutoff[index] <= 0.1:
                    if presenceRatio[index] >= 0.9:
                        if firingRate[index] >= 0.2:
                            if isiViol[index] <= 0.5:
                                date = str(session.date)
                                unit = uniqueSpikeClusters[index]
                                visualUnitsZeta[date].append(unit)
```

```python
itiSaccades = 0
sacTimes = list()
sessionDates = list()
for session in sessionsMlati:
    sessionDates.append(session.date)
sponLists = {str(session):list() for session in set(sessionDates)}
print(sponLists)
for session in sessionsMlati: 
    start = session.load('stimuli/dg/grating/timestamps')
    stop = session.load('stimuli/dg/iti/timestamps')
    saccades = session.saccadeTimestamps[:, 0]
    for i, stopTime in enumerate(stop):
        try:
            startTime = start[i + 1]
        except:
            continue
        mask = np.logical_and(saccades > stopTime, saccades < startTime)
        itiSac = len(saccades[mask])
        itiSaccades = itiSaccades + itiSac
        masked = saccades[mask]
        if masked.any():
            masked = list(masked)
            for element in masked:
                sponLists[str(session.date)].append(element)
```

```python
def pullSessionData(session):
    waveforms = session.load('saccades/predicted/left/waveforms')
    epochs = session.load('saccades/predicted/left/epochs')
    pose = session.load('pose/filtered')
    frameTimes = session.load('frames/left/timestamps')
    indices = session.load('saccades/predicted/left/indices')
    labels = session.load('saccades/predicted/left/labels')
    return waveforms, epochs, pose, frameTimes, indices, labels
```

```python
def getSaccadeFeatures(session, indices, frameTimes):
    ampList = list()
    startList = list()
    endList = list()
    FRlist = np.zeros((len(premotorUnitsZeta[str(session.date)]), len(indices)))
    agree = list()
    unitsP = premotorUnitsZeta[str(session.date)]
    unitsV = visualUnitsZeta[str(session.date)]
    for u in unitsP:
        if u in unitsV:
            agree.append(u)
    drivenIndices = list()
    for k, sac in enumerate(session.saccadeTimestamps[:, 0]):
        if sac in sponLists[str(session.date)]:
            continue
        drivenIndices.append(k)
    for i, sac in enumerate(drivenIndices):
        startIndex = sac
        endTime = session.saccadeTimestamps[sac, 1]
        if math.isnan(endTime):
            ampList.append(0)
            startList.append(0)
            endList.append(0)
            continue
        relativeEnd = abs(frameTimes - endTime)
        endShape = np.where(relativeEnd == np.min(relativeEnd))[0].shape[0]
        if endShape == 2:
            endIndex = np.where(relativeEnd == np.min(relativeEnd))[0][0]
        else:
            endIndex = int(np.where(relativeEnd == np.min(relativeEnd))[0])
        startPoint = pose[startIndex, 0]
        endPoint = pose[endIndex, 0]
        amplitude = abs(endPoint - startPoint)
        ampList.append(amplitude)
        startList.append(startPoint)
        endList.append(endPoint)
        j = 0
        for unit in session.population:
            if unit.cluster not in premotorUnitsZeta[str(session.date)]:
                continue
            spikeTimes = unit.timestamps
            start = session.saccadeTimestamps[i, 0] - 0.3
            end = session.saccadeTimestamps[i, 0]
            mask = np.logical_and(spikeTimes > start, spikeTimes < end)
            activity = len(spikeTimes[mask])/0.3
            FRlist[j, i] = activity
            j = j+1
    return(FRlist, ampList, startList, endList, drivenIndices)
```

```python
for session in sessionsMlati:
    waveforms, epochs, pose, frameTimes, indices, labels = pullSessionData(session)
    FRlist, ampList, startList, endList, drivenIndices = getSaccadeFeatures(session, indices, frameTimes)
    z = stats.zscore(FRlist, axis=1, nan_policy='omit')
    labelsDriven = labels[drivenIndices]
    sacMetrics = np.zeros((len(ampList), 4))
    sacMetrics[:, 0] = ampList
    sacMetrics[:, 1] = startList
    sacMetrics[:, 2] = endList
    sacMetrics[:, 3] = labelsDriven
   # predictedArray, predictedArray2, scoreList, scoreList2 = decodeSaccadeFeatures(session, labelsDriven, ampList, startList, endList)
   # session.save('regression/all/prediction/contra', predictedArray)
   # session.save('regression/all/prediction/ipsi', predictedArray2)
   # session.save('regression/all/r2/contra', np.array(scoreList))
   # session.save('regression/all/r2/ipsi', np.array(scoreList2))
    session.save('saccades/predicted/left/features', sacMetrics)
```
